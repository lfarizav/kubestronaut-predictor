# üè† House Data Exploration
This notebook performs initial data exploration on the `house_data.csv` dataset.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('../data/raw/kubestronaut-prediction-data.csv')

# ‚ÑπÔ∏è Display basic information about the dataset
print(f"üìê Dataset shape (rows, columns): {df.shape}")
print("\nüìä Dataset Info:")
print(df.info())
```

```python
# ‚ùì Checking for missing values to identify data quality issues
print("\nüîç Missing values per column:")
print(df.isnull().sum())
```

```python
# üìà Descriptive statistics like mean, std deviation, min, max, etc.
print("\nüìã Statistical Summary:")
print(df.describe())
```

```python
# üí∞ Visualizing the distribution of house prices
plt.figure(figsize=(10, 6))
sns.histplot(df['theory_hours'], kde=True)
plt.title('Kubestronaut-theory-hours Distribution')
plt.xlabel('theory_hours')
plt.ylabel('Frequency')
plt.show()

print("üìä Plotted kubestronaut theory hours distribution.")
```

```python
# üîó Checking how features are correlated with each other
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()

print("üìà Displayed feature correlation heatmap.")
```

```python
# üìè Scatter plot to observe relationship between sqft and price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='lab_hours', y='theory_hours', data=df)
plt.title('Theory hours vs Laboratory hours')
plt.xlabel('Laboratory hours')
plt.ylabel('Theory hours')
plt.show()

print("üìê Analyzed relation between square footage and price.")
```

```python
# üìç Bar chart to see experience years in kubernetes
plt.figure(figsize=(10, 6))
sns.countplot(x='k8s_experience_level', data=df)
plt.title('People with experience in Kubernetes')
plt.xlabel('Years of Kubernetes Experience')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

print("üìç Visualized distribution of people with years of kubernetes experience.")
```

```python
# üõèÔ∏è Bar chart showing the exam_attempts with Kubernetes experience level
plt.figure(figsize=(10, 7))
sns.barplot(x='exam_attempts', y='k8s_experience_level', data=df)
plt.title('Exam attempts with Kubernetes experience level')
plt.xlabel('exam_attempts')
plt.ylabel('Kubernetes Experience Level')
plt.show()

print("Compared Exam attempts with Kubernetes experience level.")
```
