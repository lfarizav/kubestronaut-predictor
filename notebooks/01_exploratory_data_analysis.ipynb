# üè† House Data Exploration
This notebook performs initial data exploration on the `house_data.csv` dataset.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('../data/raw/kubestronaut_predictor_data.csv')

# ‚ÑπÔ∏è Display basic information about the dataset
print(f"üìê Dataset shape (rows, columns): {df.shape}")
print("\nüìä Dataset Info:")
print(df.info())
```

```python
# ‚ùì Checking for missing values to identify data quality issues
print("\nüîç Missing values per column:")
print(df.isnull().sum())
```

```python
# üìà Descriptive statistics like mean, std deviation, min, max, etc.
print("\nüìã Statistical Summary:")
print(df.describe())
```

```python
# üí∞ Visualizing the distribution of kubestronaut final result
plt.figure(figsize=(10, 6))
sns.histplot(df['final_result'], kde=True)
plt.title('Kubestronaut final result Distribution')
plt.xlabel('final_result')
plt.ylabel('Frequency')
plt.show()

print("üìä Plotted Kubestronaut predictor distribution.")
```

```python
# üîó Checking how features are correlated with each other
plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()

print("üìà Displayed feature correlation heatmap.")
```

```python
# üìè Scatter plot to observe relationship theory hours and final result
plt.figure(figsize=(10, 6))
sns.scatterplot(x='theory_hours', y='final_result', data=df)
plt.title('Final result vs Theory hours')
plt.xlabel('Theory_hours')
plt.ylabel('Final_result')
plt.show()

print("üìê Analyzed relation between square footage and price.")
```

```python
# üìè Scatter plot to observe relationship between laboratory hours and final result
plt.figure(figsize=(10, 6))
sns.scatterplot(x='lab_hours', y='final_result', data=df)
plt.title('Final result vs Laboratory hours')
plt.xlabel('Laboratory hours')
plt.ylabel('Final result')
plt.show()

print("üìê Analyzed relation between square footage and price.")
```

```python
# üìç Bar chart to see how many people exist in each location
plt.figure(figsize=(10, 6))
sns.countplot(x='location', data=df)
plt.title('People by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

print("üìç Visualized distribution of people by location.")
```

```python
# üõèÔ∏è Bar chart showing how average final result varies by number of full exams done
plt.figure(figsize=(10, 6))
sns.barplot(x='number_full_exam_done', y='final_result', data=df)
plt.title('Average final_result by number_full_exam_done')
plt.xlabel('number_full_exam_done')
plt.ylabel('Average final_result')
plt.show()

print("üõèÔ∏è Compared average result varies by number of full exams done.")
```
